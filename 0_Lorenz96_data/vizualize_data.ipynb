```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.load("/Users/sophieabramian/Documents/GentineLabSummer2025/0_Lorenz96_data/lorenz96_N5_F8_t10000.npy")
x.shape
```

```python
timesteps = np.arange(x.shape[0])

fig = plt.figure(figsize=(7,5))
ax = fig.add_subplot(projection="3d")

# Scatter plot with time-based color
sc = ax.scatter(x[:, 0], x[:, 1], x[:, 2], c=timesteps, cmap='viridis')

ax.set_xlabel("$x_1$")
ax.set_ylabel("$x_2$")
ax.set_zlabel("$x_3$")

# Optional: colorbar to show time progression
plt.colorbar(sc, ax=ax, label="Time step", orientation='horizontal', aspect=110, shrink=0.5, pad=0.1)
plt.tight_layout()
plt.show()

```
